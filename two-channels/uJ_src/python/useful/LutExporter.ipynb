```python
from pylab import *
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def reverse_colourmap(cmap, name = 'my_cmap_r'):
    """
    In: 
    cmap, name 
    Out:
    my_cmap_r

    Explanation:
    t[0] goes from 0 to 1
    row i:   x  y0  y1 -> t[0] t[1] t[2]
                   /
                  /
    row i+1: x  y0  y1 -> t[n] t[1] t[2]

    so the inverse should do the same:
    row i+1: x  y1  y0 -> 1-t[0] t[2] t[1]
                   /
                  /
    row i:   x  y1  y0 -> 1-t[n] t[2] t[1]
    """        
    reverse = []
    k = []   

    for key in cmap._segmentdata:    
        k.append(key)
        channel = cmap._segmentdata[key]
        data = []

        for t in channel:                    
            data.append((1-t[0],t[2],t[1]))            
        reverse.append(sorted(data))    

    LinearL = dict(zip(k,reverse))
    my_cmap_r = mpl.colors.LinearSegmentedColormap(name, LinearL) 
    return my_cmap_r
```

```python


lut="autumn"


cmap = cm.get_cmap(lut, 255)    # PiYG

cmap=reverse_colourmap(cmap)

txtmap="Index\tRed\tGreen\tBlue"
for i in range(cmap.N+1):
    rgb = cmap(i)[:3] # will return rgba, we take only first 3 so we get rgb
    linemap="%s\t%s\t%s\t%s"%(i, int(rgb[0]*cmap.N),int(rgb[1]*cmap.N),int(rgb[2]*cmap.N))
    
    #hexmap=matplotlib.colors.rgb2hex(rgb)
    #linemap=("%s %s %s %s"%(i, hexmap[1:3],hexmap[3:5],hexmap[5:7]))
    txtmap=("%s\n%s"%(txtmap,linemap))

file = open("../luts/"+lut+".lut","w")  
file.write(txtmap) 
file.close() 
print(txtmap)
```
